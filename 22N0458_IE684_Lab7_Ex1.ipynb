Code for answer 4

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(xtrain,ytrain)
```

Code for answer 5

```python
l=['X1', 'X2', 'X3', 'X4', 'X5', 'X6']
```

```python
print("Intercept of trained model:",reg.intercept_)
```

```python
print('Coefficients of trained model\n')
print('Feature \t Corresponding Weight')
for i in range(6):
  print(l[i],'\t\t',reg.coef_[i])
```

Code for answer 7

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print('Mean Squared Error:',mean_squared_error(ytest,ypred))
print('R Squared Error:',r2_score(ytest,ypred))
```

Code for answer 9

```python
traintestsplit=[0.8, 0.6, 0.7, 0.9]
lambdas_list=[0.001,0.005,0.01,0.05,0.1,0.5]
```

```python
from sklearn import linear_model
for i in traintestsplit:
  for j in lambdas_list:
    print('\nTrain-Test Split:', round(i*100), ':', round((1-i)*100))
    print("Lambda =",j)
    house_train=df.sample(frac=i,random_state=100)
    house_test=df.drop(house_train.index)
    xtrain=house_train.iloc[:,:-1]
    ytrain=house_train.iloc[:,-1]
    xtest=house_test.iloc[:,:-1]
    ytest=house_test.iloc[:,-1]
    reg = linear_model.Lasso(alpha=j)
    reg.fit(xtrain,ytrain)
    print("\nIntercept:",reg.intercept_)
    print('\nCoefficients of trained model\n')
    print('Feature \t Corresponding Weight')
    for k in range(6):
      print(l[k],'\t\t',reg.coef_[k])
    ypred=reg.predict(xtest)
    print('\nMean Squared Error:',mean_squared_error(ytest,ypred))
    print('R Squared Error:',r2_score(ytest,ypred))
    print('.........................................................................................................')
```
