```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram
```

```python
df=pd.read_csv('hierarchical.csv',header=None)
```

```python
df.columns=df.iloc[0]
df=df[1:]
```

```python
df
```

```python
sc = StandardScaler()
X = sc.fit_transform(df)
#X = normalize(X)
```

```python
#pca = PCA(n_components = 2)
#X = pca.fit_transform(X)
```

```python
X
```

Ward

```python
plt.title('Dendrogram using ward method')
Dendrogram = shc.dendrogram((shc.linkage(X, method ='ward')))
```

```python
k = [2,3,4,5,6]

silhouette_scores = []
silhouette_scores.append(silhouette_score(X, AgglomerativeClustering(n_clusters = 2, linkage='ward').fit_predict(X)))
silhouette_scores.append(silhouette_score(X, AgglomerativeClustering(n_clusters = 3, linkage='ward').fit_predict(X)))
silhouette_scores.append(silhouette_score(X, AgglomerativeClustering(n_clusters = 4, linkage='ward').fit_predict(X)))
silhouette_scores.append(silhouette_score(X, AgglomerativeClustering(n_clusters = 5, linkage='ward').fit_predict(X)))
silhouette_scores.append(silhouette_score(X, AgglomerativeClustering(n_clusters = 6, linkage='ward').fit_predict(X)))


plt.bar(k, silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte scores')
plt.title('Linkage ward')
plt.show()
```

```python
ax = plt.axes(projection ="3d")
 
ax.scatter3D(X[:,0], X[:,1], X[:,2], c=AgglomerativeClustering(n_clusters = 3, linkage='ward').fit_predict(X), cmap='rainbow')
plt.title("K=3 using ward linkage")

plt.show()
```
