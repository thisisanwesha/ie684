```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def G_h(z):
    return max(0, 1-z)
```

```python
def G_l(z):
    return math.log(1+np.exp(-z))
```

```python
def G_sh(z):
    return (max(0, 1-z))**2
```

```python
z_arr=np.linspace(-5,5,100000)
```

```python
G_h_arr = np.array([G_h(z) for z in z_arr])
G_l_arr = np.array([G_l(z) for z in z_arr])
G_sh_arr = np.array([G_sh(z) for z in z_arr])
```

```python
plt.plot(z_arr, G_h_arr, label = "Hinge", color='blue')
plt.legend()
plt.title("Plot of different loss functions")
plt.xlabel('z')
plt.ylabel('Function value')
```

```python
plt.plot(z_arr, G_l_arr, label = "Logistic", color='orange')
plt.legend()
plt.title("Plot of different loss functions")
plt.xlabel('z')
plt.ylabel('Function value')
```

```python
plt.plot(z_arr, G_sh_arr, label = "Squared Hinge", color='green')
plt.legend()
plt.title("Plot of different loss functions")
plt.xlabel('z')
plt.ylabel('Function value')
```

```python
plt.plot(z_arr, G_h_arr, label = "Hinge", color='blue')
plt.plot(z_arr, G_l_arr, label = "Logistic", color='orange')
plt.plot(z_arr, G_sh_arr, label = "Squared Hinge", color='green')
plt.legend()
plt.title("Plot of different loss functions")
plt.xlabel('z')
plt.ylabel('Function value')
```

Comment: Logistic and Squared Hinge functions are continuous and differentiable but Hinge function is not differentiable at 1 althought it is also continuous.
